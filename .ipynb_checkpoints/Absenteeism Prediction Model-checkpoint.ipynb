```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
data_preprocessed = pd.read_csv('Absenteeism_preprocessed.csv')
data_preprocessed.head(10)
```

#### The model will predict excessive absenteeism, for the purpose of this demonstration of  a logistic regression we will assume that anything above the median is excessive

```python
med = data_preprocessed['Absenteeism Time in Hours'].median()
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] > med,1,0)
data_preprocessed['Excessive Absenteeism'] = targets
```

```python
data_preprocessed.head(5)
```

#### Check class balance

```python
class_dist = Counter(targets)

plt.bar(class_dist.keys(),class_dist.values())
plt.xticks(list(class_dist.keys()),
           [k for k in class_dist.keys()],
           rotation=0
           )
plt.title('Class Distribution')
plt.show()
print('Target : Percentage')
for k in class_dist.keys():
    print(f'{k} : {class_dist[k]/len(targets)*100:.2f}%')
```

 The class ratio in reasonably balanced for this demonstration.

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours', 'Month',
                                            'Daily Work Load Average', 'Distance to Work'], axis=1)

# featuers chosen to drop based on first run of model
```

```python
data_with_targets.head()
```

```python
unscaled_inputs = data_with_targets.iloc[:,:-1]
```

#### Create scaler that omits scaling dummy variables for later interpretability

```python
from sklearn.preprocessing import StandardScaler
from sklearn.base import BaseEstimator, TransformerMixin

class CustomScaler(BaseEstimator, TransformerMixin):
    
    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        self.scaler = StandardScaler(copy, with_mean, with_std)
        self.columns = columns
        self.mean_ = None
        self.std_ = None
        
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.std_ = np.std(X[self.columns])
        return self
    
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_scaled, X_not_scaled], axis=1)[init_col_order]
```

```python
unscaled_inputs.columns.values
```

```python
cols_to_omit = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Education']

cols_to_scale = [x for x in unscaled_inputs.columns.values if x not in cols_to_omit]
```

```python
absenteeism_scaler = CustomScaler(cols_to_scale)
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
```

```python
scaled_inputs
```

```python
scaled_inputs.shape
```

#### Create Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
#  split data 
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size=0.8, random_state=42)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
from sklearn import set_config
set_config(print_changed_only=False)
```

```python
reg = LogisticRegression()
reg.fit(x_train,y_train)
```

```python
outputs = reg.predict(x_train)
correct_sum = sum((outputs == y_train))
score = correct_sum/outputs.shape[0]
```

```python
print(score)
```

```python
feature_names = unscaled_inputs.columns.values

summary_table = pd.DataFrame(columns=['Feature Name'], data=feature_names)
summary_table['Coefficients'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index +1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

```python
summary_table['Odds Ratio'] = np.exp(summary_table.Coefficients)
summary_table.sort_values('Odds Ratio', ascending=False)
```

### Test Data
Since this demonstration focussed on integrating SQL, Python and Tableau. A simple score and probabilities will be produced.

```python
reg.score(x_test, y_test)
```
